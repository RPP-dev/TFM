```python
# PARTE 1 - Exploración de Datos

# Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar dataset
df = pd.read_csv('dataset.csv')
```

```python
# Limpieza inicial
df.columns = df.columns.str.strip()  # Quitar espacios
print(df.info())
print(df.describe())

# Eliminar edades negativas
df = df[df['Age'] >= 0]
```

```python
# Análisis Univariante
plt.figure(figsize=(10,6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Distribución de Edades')
plt.xlabel('Edad')
plt.ylabel('Cantidad')
plt.show()

# Análisis de Ausencias
no_show_counts = df['No-show'].value_counts()
print(no_show_counts)
```

```python
plt.figure(figsize=(6,6))
no_show_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Proporción de No-show')
plt.show()

# Análisis Bivariante
plt.figure(figsize=(8,6))
sns.countplot(x='Gender', hue='No-show', data=df)
plt.title('Relación Género y Asistencia')
plt.show()
```

```python
# Guardar dataset limpio
df.to_csv('dataset.csv', index=False)
```
